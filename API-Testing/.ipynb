####Â Historical Market Data (Number of Days)
Get historical market data include price, market cap, and 24h volume (granularity auto) for a **specific coin**.

Minutely data will be used for duration within 1 day, Hourly data will be used for duration between 1 day and 90 days, Daily data will be used for duration above 90 days.



**Get historical data of specific coins and write data in a data frame**

```python
coins = ["convex-finance", "ribbon-finance", "rari-governance-token", "gmx", "nftx", "ftx-token", "spell-token", "dopex" , "gro-dao-token", "audius", "ecomi", "raydium"]

column_names = ["prices", "market_caps", "total_volumes", "coin"]
df_coin_historical_data = pd.DataFrame(columns=column_names)

for coin in coins:
    url = 'https://api.coingecko.com/api/v3/coins/' + coin + '/market_chart?vs_currency=usd&days=100'
    r = requests.get(url)
    r.json()
    df_coin = pd.DataFrame(r.json())    
    df_coin["coin"] = coin
    df_coin_historical_data = df_coin_historical_data.append(df_coin, ignore_index=True)


print(df_coin_historical_data)
```

####Â Historical Market Data (Daterange)
Get historical market data include price, market cap, and 24h volume within a range of timestamp (granularity auto) for a **specific coin**.

* Data granularity is automatic (cannot be adjusted)
* 1 day from query time = 5 minute interval data
* 1 - 90 days from query time = hourly data
* above 90 days from query time = daily data (00:00 UTC)

Parmeter *from* and *to* have to be in UNIX Timestamp

```python
import datetime
import time
from_date = datetime.datetime(2020, 1, 1) 
from_date = int((time.mktime(from_date.timetuple())))
print()
print("Unix Timestamp from: ",from_date)
print()

to_date = datetime.datetime(2021, 10, 9)
to_date = int((time.mktime(to_date.timetuple())))
print("Unix Timestamp to: ",to_date)
print()
```
